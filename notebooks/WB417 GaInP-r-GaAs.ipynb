# Modeled 3T GaInP/s/GaAs (WB417) performance from SI of:
### Geisz, et al., Cell Reports Physical Science __2__, 100677 (2021). 

```python
WB417 = pvc.Tandem3T()
```

```python
WB417.set(name='WB417', Rz = 2.3)
WB417.top.set(Eg=1.87, J0ratio=[ 80., 22.], Jext=0.0131, Gsh=1e-8)   
WB417.bot.set(Eg=1.419, J0ratio=[10., 15.], Jext=0.0128, Gsh= 5e-5)
WB417.bot.set(Rser=0.2, beta=5., area=0.89,  RBB='JFG')
```

```python
WB417.controls()
```

```python
pnts=51
```

```python
CZ_figs, CZ_axs, CZ_iv, CZ_sp = WB417.plot(meastype='CZ', pnts=pnts)
```

```python
CR_figs, CR_axs, CR_iv, CR_sp = WB417.plot(meastype='CR', pnts=pnts)
```

```python
CT_figs, CT_xs, CT_iv, CT_sp = WB417.plot(meastype='CT', pnts=pnts)
```
