# 3T GaInP/s/GaAs (WB417) results from SI of CRPS paper

```python
WB417 = pvc.Tandem3T()
```

```python
WB417.update(name='WB417', Rz = 2.3)
WB417.top.update(Eg=1.87, J0ratio=[ 80., 22.], Jext=0.0131, Gsh=1e-8)   
WB417.bot.update(Eg=1.419, J0ratio=[10., 15.], Jext=0.0128, Gsh= 5e-5)
WB417.bot.update(Rser=0.2, beta=5., area=0.89,  RBB='JFG')
```

```python
WB417
```

```python
pnts=101
```

```python
CZV_fig, CZI_fig, CZ_axs, CZ_iv, CZ_sp = WB417.plot(meastype='CZ', pnts=pnts)
```

```python
CR_Vfig, CR_Ifig, CR_axs, CR_iv, CR_sp = WB417.plot(meastype='CR', pnts=pnts)
```

```python
CT_Vfig, CT_Ifig, CT_xs, CT_iv, CT_sp = WB417.plot(meastype='CT', pnts=pnts)
```

```python
hex_Vfig, hex_Ifig, hex_axs, hex_iv, hex_sp = WB417.plot(meastype='CZ',oper='dev2hex', pnts=pnts)
```

```python
hex_axs[0].set_xlim(-1,3)
hex_axs[0].set_ylim(-2,1.5)
hex_axs[1].set_xlim(-25,5)
hex_axs[1].set_ylim(-35,25)
```
